# Laboratorio: Minería de Datos — Regresión Lineal Múltiple
**Metodología:** CRISP-DM  
**Modelo:** `sklearn.linear_model.LinearRegression`  
**Ejercicios:** Precio del Dólar · Glucosa en Sangre · Consumo de Energía

````python
# ─────────────────────────────────────────────
# IMPORTS GLOBALES
# ─────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Estilo global de gráficas
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 110
````

---
# EJERCICIO 1 — Predicción del Precio del Dólar
**Dataset:** `dolar_data.csv`  
**Variable dependiente:** `Precio_Dolar`  
**Predictores:** `Dia`, `Inflacion`, `Tasa_interes`

## Fase 1 — Business Understanding
Se busca predecir el precio del dólar a partir de indicadores macroeconómicos diarios.
Esto permite anticipar movimientos cambiarios usando únicamente variables fácilmente observables.

````python
# ─────────────────────────────────────────────
# Fase 2 — DATA UNDERSTANDING
# Carga y exploración inicial del dataset
# ─────────────────────────────────────────────
df_dolar = pd.read_csv('dolar_data.csv')

print('Shape:', df_dolar.shape)
print('\nPrimeras filas:')
display(df_dolar.head())

print('\nEstadísticas descriptivas:')
display(df_dolar.describe())

print('\nValores nulos por columna:')
print(df_dolar.isnull().sum())
````

````python
# ─────────────────────────────────────────────
# Fase 2 — Matriz de correlación
# Nos muestra qué tan relacionadas están las variables entre sí
# ─────────────────────────────────────────────
plt.figure(figsize=(7, 5))
sns.heatmap(df_dolar.corr(), annot=True, fmt='.3f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlación — Dataset Dólar')
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Fase 3 — DATA PREPARATION
# Separación de variables y división train/test (80/20)
# ─────────────────────────────────────────────
X1 = df_dolar[['Dia', 'Inflacion', 'Tasa_interes']]
y1 = df_dolar['Precio_Dolar']

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42
)

print(f'Train: {X1_train.shape[0]} filas | Test: {X1_test.shape[0]} filas')
````

````python
# ─────────────────────────────────────────────
# Fase 4 — MODELING
# Entrenamiento del modelo de Regresión Lineal Múltiple
# Documentación: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
# ─────────────────────────────────────────────
modelo_dolar = LinearRegression()
modelo_dolar.fit(X1_train, y1_train)

# Ecuación del modelo: Precio_Dolar = b0 + b1*Dia + b2*Inflacion + b3*Tasa_interes
print('Intercepto (β₀):', round(modelo_dolar.intercept_, 4))
print('\nCoeficientes:')
for col, coef in zip(X1.columns, modelo_dolar.coef_):
    print(f'  {col:20s}: {coef:.4f}')
````

### Interpretación de coeficientes — Ejercicio 1

| Variable | Coeficiente | Interpretación |
|---|---|---|
| `Dia` | β₁ | Por cada día adicional, el precio del dólar **aumenta en β₁ unidades** (tendencia temporal positiva) |
| `Inflacion` | β₂ | Un aumento de 1 unidad en inflación diaria **modifica** el precio del dólar en β₂ (generalmente positivo: más inflación → dólar más caro) |
| `Tasa_interes` | β₃ | Por cada punto adicional en la tasa de interés, el precio varía en β₃ (efecto puede ser negativo: tasas altas atraen capital y aprecian la moneda local) |

> **Nota:** Los valores exactos se imprimen en la celda anterior. La magnitud relativa indica la importancia de cada variable.

````python
# ─────────────────────────────────────────────
# Fase 5 — EVALUATION
# Métricas de desempeño: MSE y R²
# ─────────────────────────────────────────────
y1_pred = modelo_dolar.predict(X1_test)

mse1  = mean_squared_error(y1_test, y1_pred)
rmse1 = np.sqrt(mse1)
r2_1  = r2_score(y1_test, y1_pred)

print('═══ DESEMPEÑO DEL MODELO — DÓLAR ═══')
print(f'MSE  : {mse1:.4f}   (Error cuadrático medio: menor es mejor)')
print(f'RMSE : {rmse1:.4f}  (Error en las mismas unidades que Precio_Dolar)')
print(f'R²   : {r2_1:.4f}   (1.0 = ajuste perfecto | 0 = sin poder predictivo)')
````

````python
# ─────────────────────────────────────────────
# Fase 5 — Gráfica: Predicho vs Real
# ─────────────────────────────────────────────
plt.figure(figsize=(6, 5))
plt.scatter(y1_test, y1_pred, alpha=0.6, edgecolors='k', linewidth=0.4)
lims = [min(y1_test.min(), y1_pred.min()), max(y1_test.max(), y1_pred.max())]
plt.plot(lims, lims, 'r--', label='Predicción perfecta')
plt.xlabel('Precio Real')
plt.ylabel('Precio Predicho')
plt.title('Real vs Predicho — Precio del Dólar')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Fase 5 — Visualización: relación de cada predictor con Precio_Dolar
# ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
predictores1 = ['Dia', 'Inflacion', 'Tasa_interes']

for ax, col in zip(axes, predictores1):
    ax.scatter(df_dolar[col], df_dolar['Precio_Dolar'], alpha=0.4, s=15)
    # Línea de tendencia
    m, b = np.polyfit(df_dolar[col], df_dolar['Precio_Dolar'], 1)
    x_line = np.linspace(df_dolar[col].min(), df_dolar[col].max(), 100)
    ax.plot(x_line, m * x_line + b, 'r-', linewidth=2)
    ax.set_xlabel(col)
    ax.set_ylabel('Precio_Dolar')
    ax.set_title(f'{col} vs Precio_Dolar')

plt.suptitle('Relación de Variables Independientes con el Precio del Dólar', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Exportación del modelo con joblib
# joblib es más eficiente que pickle para arrays NumPy grandes
# ─────────────────────────────────────────────
joblib.dump(modelo_dolar, 'modelo_dolar.pkl')
print('Modelo guardado como: modelo_dolar.pkl')

# Verificación: cargar y probar
modelo_cargado = joblib.load('modelo_dolar.pkl')
test_pred = modelo_cargado.predict([[100, 0.02, 5.0]])
print(f'Predicción de prueba (Dia=100, Inflacion=0.02, Tasa=5.0): {test_pred[0]:.2f}')
````

# EJERCICIO 2 — Predicción de Niveles de Glucosa
**Dataset:** `glucosa_data.csv`  
**Variable dependiente:** `Nivel_Glucosa` (mg/dL)  
**Predictores:** `Edad`, `IMC`, `Actividad_Fisica`

## Fase 1 — Business Understanding
El nivel de glucosa en sangre es un indicador clave de la salud metabólica.
Predecirlo a partir de variables clínicas simples (edad, IMC, actividad física) 
permite identificar pacientes en riesgo de diabetes sin necesidad de exámenes complejos.

````python
# ─────────────────────────────────────────────
# Fase 2 — DATA UNDERSTANDING
# ─────────────────────────────────────────────
df_glucosa = pd.read_csv('glucosa_data.csv')

print('Shape:', df_glucosa.shape)
print('\nPrimeras filas:')
display(df_glucosa.head())

print('\nEstadísticas descriptivas:')
display(df_glucosa.describe())

print('\nValores nulos:')
print(df_glucosa.isnull().sum())
````

````python
# ─────────────────────────────────────────────
# Fase 2 — Matriz de correlación
# ─────────────────────────────────────────────
plt.figure(figsize=(7, 5))
sns.heatmap(df_glucosa.corr(), annot=True, fmt='.3f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlación — Dataset Glucosa')
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Fase 3 — DATA PREPARATION
# División train/test 80/20
# ─────────────────────────────────────────────
X2 = df_glucosa[['Edad', 'IMC', 'Actividad_Fisica']]
y2 = df_glucosa['Nivel_Glucosa']

X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.2, random_state=42
)
print(f'Train: {X2_train.shape[0]} filas | Test: {X2_test.shape[0]} filas')
````

````python
# ─────────────────────────────────────────────
# Fase 4 — MODELING
# ─────────────────────────────────────────────
modelo_glucosa = LinearRegression()
modelo_glucosa.fit(X2_train, y2_train)

print('Intercepto (β₀):', round(modelo_glucosa.intercept_, 4))
print('\nCoeficientes:')
coef_df2 = pd.DataFrame({
    'Variable': X2.columns,
    'Coeficiente': modelo_glucosa.coef_,
    'Abs_Coef': np.abs(modelo_glucosa.coef_)
}).sort_values('Abs_Coef', ascending=False)
display(coef_df2)
````

### Interpretación de coeficientes — Ejercicio 2

| Variable | Coeficiente | Interpretación |
|---|---|---|
| `Edad` | β₁ | Por cada año adicional de edad, el nivel de glucosa **aumenta en β₁ mg/dL** (el metabolismo glucídico empeora con la edad) |
| `IMC` | β₂ | Un punto más de IMC eleva la glucosa en β₂ mg/dL (el exceso de grasa corporal reduce la sensibilidad a la insulina) |
| `Actividad_Fisica` | β₃ | Cada hora semanal adicional de ejercicio **reduce** la glucosa en |β₃| mg/dL (efecto protector del ejercicio) |

> La variable con **mayor valor absoluto de coeficiente** es la de mayor impacto. Observar la tabla `coef_df2` para determinar cuál domina.

````python
# ─────────────────────────────────────────────
# Fase 5 — EVALUATION
# ─────────────────────────────────────────────
y2_pred = modelo_glucosa.predict(X2_test)

mse2  = mean_squared_error(y2_test, y2_pred)
rmse2 = np.sqrt(mse2)
r2_2  = r2_score(y2_test, y2_pred)

print('═══ DESEMPEÑO DEL MODELO — GLUCOSA ═══')
print(f'MSE  : {mse2:.4f}')
print(f'RMSE : {rmse2:.4f}  (en mg/dL)')
print(f'R²   : {r2_2:.4f}')
````

````python
# ─────────────────────────────────────────────
# Fase 5 — Gráfica: Predicho vs Real
# ─────────────────────────────────────────────
plt.figure(figsize=(6, 5))
plt.scatter(y2_test, y2_pred, alpha=0.4, edgecolors='k', linewidth=0.3, s=20)
lims = [min(y2_test.min(), y2_pred.min()), max(y2_test.max(), y2_pred.max())]
plt.plot(lims, lims, 'r--', label='Predicción perfecta')
plt.xlabel('Glucosa Real (mg/dL)')
plt.ylabel('Glucosa Predicha (mg/dL)')
plt.title('Real vs Predicho — Nivel de Glucosa')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Fase 5 — Visualización: relación de cada predictor con Nivel_Glucosa
# ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
predictores2 = ['Edad', 'IMC', 'Actividad_Fisica']

for ax, col in zip(axes, predictores2):
    ax.scatter(df_glucosa[col], df_glucosa['Nivel_Glucosa'], alpha=0.3, s=12)
    m, b = np.polyfit(df_glucosa[col], df_glucosa['Nivel_Glucosa'], 1)
    x_line = np.linspace(df_glucosa[col].min(), df_glucosa[col].max(), 100)
    ax.plot(x_line, m * x_line + b, 'r-', linewidth=2)
    ax.set_xlabel(col)
    ax.set_ylabel('Nivel_Glucosa (mg/dL)')
    ax.set_title(f'{col} vs Nivel_Glucosa')

plt.suptitle('Relación de Variables Independientes con el Nivel de Glucosa', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Análisis de importancia de variables
# Usamos coeficientes estandarizados para comparar "manzanas con manzanas"
# ─────────────────────────────────────────────
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X2_scaled = scaler2.fit_transform(X2_train)
modelo_glucosa_std = LinearRegression()
modelo_glucosa_std.fit(X2_scaled, y2_train)

importancia2 = pd.DataFrame({
    'Variable': X2.columns,
    'Coef_Estandarizado': modelo_glucosa_std.coef_
}).assign(Abs=lambda d: d['Coef_Estandarizado'].abs())\
  .sort_values('Abs', ascending=False)

print('Variables ordenadas por impacto (coeficientes estandarizados):')
display(importancia2.drop(columns='Abs'))

# Barplot de importancia
plt.figure(figsize=(6, 3))
colors = ['steelblue' if c > 0 else 'tomato' for c in importancia2['Coef_Estandarizado']]
plt.barh(importancia2['Variable'], importancia2['Coef_Estandarizado'], color=colors)
plt.axvline(0, color='black', linewidth=0.8)
plt.xlabel('Coeficiente Estandarizado')
plt.title('Impacto de cada variable en Nivel_Glucosa')
plt.tight_layout()
plt.show()
````

````python
# Exportar modelo de glucosa
joblib.dump(modelo_glucosa, 'modelo_glucosa.pkl')
print('Modelo guardado como: modelo_glucosa.pkl')

# Verificación
modelo_glc = joblib.load('modelo_glucosa.pkl')
test_pred2 = modelo_glc.predict([[45, 25.0, 3]])
print(f'Predicción de prueba (Edad=45, IMC=25, Act=3h): {test_pred2[0]:.2f} mg/dL')
````

# EJERCICIO 3 — Predicción del Consumo de Energía Eléctrica
**Dataset:** `energia_data.csv`  
**Variable dependiente:** `Consumo_Energia` (kWh)  
**Predictores:** `Temperatura`, `Hora` (cíclica → sin/cos), `Dia_Semana`

## Fase 1 — Business Understanding
Las empresas de energía necesitan anticipar la demanda para equilibrar la red eléctrica.
La temperatura ambiente, la hora del día y el día de la semana son los principales 
conductores del consumo. Se aplica una **transformación cíclica seno/coseno** a `Hora`
porque el consumo a las 23h es más similar al de las 1h que al de las 12h 
— una escala numérica lineal no captura esa circularidad.

````python
# ─────────────────────────────────────────────
# Fase 2 — DATA UNDERSTANDING
# ─────────────────────────────────────────────
df_energia = pd.read_csv('energia_data.csv')

print('Shape:', df_energia.shape)
display(df_energia.head())
display(df_energia.describe())
print('Valores nulos:', df_energia.isnull().sum().to_dict())
````

````python
# ─────────────────────────────────────────────
# Fase 2 — Matriz de correlación
# ─────────────────────────────────────────────
plt.figure(figsize=(7, 5))
sns.heatmap(df_energia.corr(), annot=True, fmt='.3f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlación — Dataset Energía')
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Fase 3 — DATA PREPARATION
# Transformación cíclica de Hora (seno y coseno)
# Hora va de 1 a 24 → se normaliza a [0, 2π]
# ─────────────────────────────────────────────
df_energia['Hora_sin'] = np.sin(2 * np.pi * df_energia['Hora'] / 24)
df_energia['Hora_cos'] = np.cos(2 * np.pi * df_energia['Hora'] / 24)

# Visualizar la transformación cíclica
fig, ax = plt.subplots(figsize=(8, 4))
horas = np.arange(1, 25)
ax.plot(horas, np.sin(2 * np.pi * horas / 24), 'b-o', markersize=4, label='sin(Hora)')
ax.plot(horas, np.cos(2 * np.pi * horas / 24), 'r-s', markersize=4, label='cos(Hora)')
ax.set_xlabel('Hora del día')
ax.set_ylabel('Valor transformado')
ax.set_title('Transformación Cíclica de la Variable Hora (1–24)')
ax.legend()
plt.tight_layout()
plt.show()

print('Nuevas columnas:', ['Hora_sin', 'Hora_cos'])
````

````python
# ─────────────────────────────────────────────
# Fase 3 — División train/test
# Se usa Hora_sin y Hora_cos en lugar de Hora directa
# ─────────────────────────────────────────────
X3 = df_energia[['Temperatura', 'Hora_sin', 'Hora_cos', 'Dia_Semana']]
y3 = df_energia['Consumo_Energia']

X3_train, X3_test, y3_train, y3_test = train_test_split(
    X3, y3, test_size=0.2, random_state=42
)
print(f'Train: {X3_train.shape[0]} filas | Test: {X3_test.shape[0]} filas')
````

````python
# ─────────────────────────────────────────────
# Fase 4 — MODELING
# ─────────────────────────────────────────────
modelo_energia = LinearRegression()
modelo_energia.fit(X3_train, y3_train)

print('Intercepto (β₀):', round(modelo_energia.intercept_, 4))
print('\nCoeficientes:')
for col, coef in zip(X3.columns, modelo_energia.coef_):
    print(f'  {col:15s}: {coef:.4f}')
````

### Interpretación de coeficientes — Ejercicio 3

| Variable | Coeficiente | Interpretación |
|---|---|---|
| `Temperatura` | β₁ | Por cada °C adicional, el consumo cambia en β₁ kWh (positivo: más calor → más aire acondicionado) |
| `Hora_sin` | β₂ | Componente sinusoidal de la hora. Captura el pico de la tarde |
| `Hora_cos` | β₃ | Componente coseno. Captura las horas de madrugada de bajo consumo |
| `Dia_Semana` | β₄ | Diferencia entre días laborables (1–5) y fin de semana (6–7) |

> La **Temperatura** suele ser la variable de mayor impacto en consumo energético residencial.

````python
# ─────────────────────────────────────────────
# Fase 5 — EVALUATION: RMSE y R² (como pide el enunciado)
# ─────────────────────────────────────────────
y3_pred = modelo_energia.predict(X3_test)

mse3  = mean_squared_error(y3_test, y3_pred)
rmse3 = np.sqrt(mse3)
r2_3  = r2_score(y3_test, y3_pred)

print('═══ DESEMPEÑO DEL MODELO — ENERGÍA ═══')
print(f'MSE  : {mse3:.4f}')
print(f'RMSE : {rmse3:.4f}  (en kWh — error promedio de predicción)')
print(f'R²   : {r2_3:.4f}')
````

````python
# ─────────────────────────────────────────────
# Fase 5 — Gráfica: Predicho vs Real
# ─────────────────────────────────────────────
plt.figure(figsize=(6, 5))
plt.scatter(y3_test, y3_pred, alpha=0.2, s=10, edgecolors='k', linewidth=0.2)
lims = [min(y3_test.min(), y3_pred.min()), max(y3_test.max(), y3_pred.max())]
plt.plot(lims, lims, 'r--', label='Predicción perfecta')
plt.xlabel('Consumo Real (kWh)')
plt.ylabel('Consumo Predicho (kWh)')
plt.title('Real vs Predicho — Consumo de Energía')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Fase 5 — Visualizaciones de relaciones importantes
# ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1. Temperatura vs Consumo
axes[0].scatter(df_energia['Temperatura'], df_energia['Consumo_Energia'], alpha=0.15, s=8)
m, b = np.polyfit(df_energia['Temperatura'], df_energia['Consumo_Energia'], 1)
x_l = np.linspace(df_energia['Temperatura'].min(), df_energia['Temperatura'].max(), 100)
axes[0].plot(x_l, m * x_l + b, 'r-', linewidth=2)
axes[0].set_xlabel('Temperatura (°C)')
axes[0].set_ylabel('Consumo_Energia (kWh)')
axes[0].set_title('Temperatura vs Consumo')

# 2. Hora vs Consumo (boxplot por hora)
consumo_por_hora = df_energia.groupby('Hora')['Consumo_Energia'].mean()
axes[1].bar(consumo_por_hora.index, consumo_por_hora.values, color='steelblue', alpha=0.8)
axes[1].set_xlabel('Hora del día')
axes[1].set_ylabel('Consumo Promedio (kWh)')
axes[1].set_title('Consumo Promedio por Hora')

# 3. Dia_Semana vs Consumo
dias_labels = {1:'Lun',2:'Mar',3:'Mié',4:'Jue',5:'Vie',6:'Sáb',7:'Dom'}
consumo_por_dia = df_energia.groupby('Dia_Semana')['Consumo_Energia'].mean()
axes[2].bar([dias_labels[d] for d in consumo_por_dia.index], consumo_por_dia.values,
            color='coral', alpha=0.85)
axes[2].set_xlabel('Día de la Semana')
axes[2].set_ylabel('Consumo Promedio (kWh)')
axes[2].set_title('Consumo Promedio por Día')

plt.suptitle('Relaciones Variables Independientes vs Consumo de Energía', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
````

````python
# ─────────────────────────────────────────────
# Importancia de variables — coeficientes estandarizados
# ─────────────────────────────────────────────
scaler3 = StandardScaler()
X3_scaled = scaler3.fit_transform(X3_train)
modelo_energia_std = LinearRegression()
modelo_energia_std.fit(X3_scaled, y3_train)

importancia3 = pd.DataFrame({
    'Variable': X3.columns,
    'Coef_Estandarizado': modelo_energia_std.coef_
}).assign(Abs=lambda d: d['Coef_Estandarizado'].abs())\
  .sort_values('Abs', ascending=False)

print('Variable con mayor impacto en Consumo_Energia:')
print(' →', importancia3.iloc[0]['Variable'])
display(importancia3.drop(columns='Abs'))

plt.figure(figsize=(6, 3))
colors3 = ['steelblue' if c > 0 else 'tomato' for c in importancia3['Coef_Estandarizado']]
plt.barh(importancia3['Variable'], importancia3['Coef_Estandarizado'], color=colors3)
plt.axvline(0, color='black', linewidth=0.8)
plt.xlabel('Coeficiente Estandarizado')
plt.title('Impacto de cada variable en Consumo_Energia')
plt.tight_layout()
plt.show()
````

````python
# Exportar modelo de energía
joblib.dump(modelo_energia, 'modelo_energia.pkl')
print('Modelo guardado como: modelo_energia.pkl')

# Verificación
modelo_en = joblib.load('modelo_energia.pkl')
hora_test = 18
hora_sin_t = np.sin(2 * np.pi * hora_test / 24)
hora_cos_t = np.cos(2 * np.pi * hora_test / 24)
test_pred3 = modelo_en.predict([[30.0, hora_sin_t, hora_cos_t, 2]])
print(f'Predicción (Temp=30°C, Hora=18h, Martes): {test_pred3[0]:.2f} kWh')
````

````python
# ─────────────────────────────────────────────
# Tabla resumen de métricas de los 3 modelos
# ─────────────────────────────────────────────
resumen = pd.DataFrame({
    'Modelo'     : ['Precio Dólar', 'Nivel Glucosa', 'Consumo Energía'],
    'Variable Y' : ['Precio_Dolar', 'Nivel_Glucosa', 'Consumo_Energia'],
    'MSE'        : [round(mse1, 2), round(mse2, 2), round(mse3, 2)],
    'RMSE'       : [round(rmse1, 2), round(rmse2, 2), round(rmse3, 2)],
    'R²'         : [round(r2_1, 4), round(r2_2, 4), round(r2_3, 4)],
    'Modelo .pkl': ['modelo_dolar.pkl', 'modelo_glucosa.pkl', 'modelo_energia.pkl']
})

display(resumen)
print('\nInterpretación R²:')
print('  > 0.90 → Excelente | 0.70–0.90 → Bueno | 0.50–0.70 → Moderado | < 0.50 → Débil')
````

```python
joblib.dump(modelo, 'nombre_modelo.pkl')

# CARGAR en otra aplicación
modelo_cargado = joblib.load('nombre_modelo.pkl')
prediccion = modelo_cargado.predict([[valor1, valor2, valor3]])
```

**¿Por qué `joblib` y no `pickle`?**  
`joblib` está optimizado para serializar objetos con arrays NumPy internos (como los modelos de scikit-learn), es más rápido y produce archivos más pequeños. `pickle` también funciona pero es menos eficiente en este contexto.
